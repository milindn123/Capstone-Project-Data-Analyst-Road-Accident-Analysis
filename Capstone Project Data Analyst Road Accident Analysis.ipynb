```python
#1. The percentage of road accidents during all the years.

# summing the number of accident for each year
sum_of_accident_by_years = df[['State/UT-Wise Total Number of Road Accidents during - 2013', 'State/UT-Wise Total Number of Road Accidents during - 2014', 'State/UT-Wise Total Number of Road Accidents during - 2015', 'State/UT-Wise Total Number of Road Accidents during - 2016']].sum()

# Calculating the total number of accidents across all years
total_accidents_all_years = sum_of_accident_by_years.sum()

# Calculating the total percentage accident for each year
percent_accident_per_year = (sum_of_accident_by_years/total_accidents_all_years)*100

# Lets Print the Result
print(percent_accident_per_year)
```

As we can observe that the percentage of road accidents during all the years is almost same or equal to.

```python
#2. Mean Accidents per 1L population for each year.

mean_of_2013 = np.mean(df['Total Number of Accidents Per Lakh Population - 2013'])
print('Mean for the year 2013 -', mean_of_2013)

mean_of_2014 = np.mean(df['Total Number of Accidents Per Lakh Population - 2014'])
print('Mean for the year 2014 -', mean_of_2014)

mean_of_2015 = np.mean(df['Total Number of Accidents Per Lakh Population - 2015'])
print('Mean for the year 2015 -', mean_of_2015)

mean_of_2016 = np.mean(df['Total Number of Accidents Per Lakh Population - 2016'])
print('Mean for the year 2016 -', mean_of_2016)
```

```python
#The highest number of accident states and least number of accident states.

plt.figure(figsize = (20,10))
plt.rcParams.update({'font.size':18})
y = df['State/UT-Wise Total Number of Road Accidents during - 2013']
yd = df['States/UTs']
p =df['States/UTs'].nunique()
d = np.linspace(1,p,p)
plt.bar(d, y, align = 'center')
plt.xticks(d, yd, rotation = 90)
plt.xlabel('States')
plt.ylabel('Number of Accidents')
plt.title('Total number of Accidents in each State.')
plt.show()
```

Maharashtra and Tamil Nadu have the highest number of accidents. Arunachal, Manipur, Meghalaya, Mizoram, Nagaland, Tripura have the least number of accidents. They surprisingly all belong to the north-eastern area.
